# Dataloader
Loads data into the standard $\mathbf{X}y$ format.

### Import packages

```python
import torch
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt

import os
```

## Load data from .pt files to torch tensors
This function loads datasets from the 'datasets' folder, and adds the loaded tensors to dictionaries.

```python
## Define function for loading data and adding to dictionaries
def Load_Data(datasets_list):

    Xtrain_dict = {}
    ytrain_dict = {}
    Xtest_dict = {}
    ytest_dict = {}

    for dataset in datasets_list:
        
        trainData = torch.load(os.path.join('datasets', dataset, 'train.pt'))
        testData = torch.load(os.path.join('datasets', dataset, 'test.pt'))
        
        Xtrain = trainData['samples'].squeeze(1)
        ytrain = trainData['labels']
        Xtest = testData['samples'].squeeze(1)
        ytest = testData['labels']

        ## Add loaded data to dictionaries
        Xtrain_dict[dataset] = Xtrain
        ytrain_dict[dataset] = ytrain
        Xtest_dict[dataset] = Xtest
        ytest_dict[dataset] = ytest

    return Xtrain_dict, ytrain_dict, Xtest_dict, ytest_dict
```

```python
datasets_list = ['SleepEEG', 'Epilepsy', 'FD-A', 'FD-B', 'HAR', 'Gesture', 'ecg', 'emg']

## Load data
Xtrain_dict, ytrain_dict, Xtest_dict, ytest_dict = Load_Data(datasets_list)
```
