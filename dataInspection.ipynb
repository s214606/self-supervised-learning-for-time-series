Inspektion af Epilepsy

```python
from DataLoader import TimeSeriesDataset
from configs.Epilepsy_configs import Config
config = Config()
import torch
with open('datasets/Epilepsy/train.pt', 'rb') as f:
    dataset = torch.load(f)
datasetEPTrain = TimeSeriesDataset(dataset,config=config)

with open('datasets/Epilepsy/test.pt', 'rb') as f:
    dataset = torch.load(f)
datasetEPTest = TimeSeriesDataset(dataset,config=config)

with open('datasets/Epilepsy/val.pt', 'rb') as f:
    dataset = torch.load(f)
datasetEPVal = TimeSeriesDataset(dataset,config=config)

```

```python
labelsEPTrain = [int(y) for y in datasetEPTrain.y]
labelsEPTest = [int(y) for y in datasetEPTest.y]
labelsEPVal = [int(y) for y in datasetEPVal.y]
```

```python
from collections import Counter

binsEPTrain = Counter(labelsEPTrain)
procentEPTrain = [100*v/sum(binsEPTrain.values()) for v in binsEPTrain.values()]
keyEPTrain = binsEPTrain.keys()
print("Train")
print(procentEPTrain)
print(keyEPTrain)

binsEPTest = Counter(labelsEPTest)
procentEPTest = [100*v/sum(binsEPTest.values()) for v in binsEPTest.values()]
keyEPTest = binsEPTest.keys()
print("Test")
print(procentEPTest)
print(keyEPTest)

binsEPVal = Counter(labelsEPVal)
procentEPVal = [100*v/sum(binsEPVal.values()) for v in binsEPVal.values()]
keyEPVal = binsEPVal.keys()
print("Val")
print(procentEPVal)
print(keyEPVal)
```

```python
#Plot labels in a distribution plot, where labels are the classes and not continuous values
import matplotlib.pyplot as plt
from collections import Counter

bins = Counter(labelsEPTrain)
color1 = '#E83F48'
color2 = '#030F4F90'
#plot bins and include x labels for every bin
#activity_labels = ["Walking","Jogging","Stairs","Sitting","Standing","Typing","Brushing teeth","Eating soup","Eating chpis","Eating pasta","Drinking from cup","Eating sandwich","Kicking ball","Playing catch","Dribbing","Writing","Clapping","Folding clothes"]
fig = plt.figure(figsize=(5,5))
plt.bar(bins.keys(), bins.values(),color=[color2,color1])
plt.xticks(list(sorted(bins.keys())))
#set a title on the plot
plt.title("Distribution of the subjects has epelepsy or not in the train dataset")
#turn the labels 90 degrees
#plt.xticks(rotation=60)
#set axis labels
plt.xlabel("Classes in train dataset")
plt.ylabel("Number of samples in train dataset")

plt.show()
```
