```python
#Plot labels in a distribution plot, where labels are the classes and not continuous values
import matplotlib.pyplot as plt
from collections import Counter

bins = Counter(labels)
color1 = '#E83F48'
color2 = '#030F4F90'
#plot bins and include x labels for every bin
#activity_labels = ["Walking","Jogging","Stairs","Sitting","Standing","Typing","Brushing teeth","Eating soup","Eating chpis","Eating pasta","Drinking from cup","Eating sandwich","Kicking ball","Playing catch","Dribbing","Writing","Clapping","Folding clothes"]
fig = plt.figure(figsize=(10,5))
plt.bar(bins.keys(), sorted([100*v/sum(bins.values()) for v in bins.values()]),color=[color1,color2,color1,color2,color1])
plt.xticks(list(bins.keys()))
#set a title on the plot
plt.title("Distribution of activities in the dataset")
#turn the labels 90 degrees
#plt.xticks(rotation=60)
#set axis labels
plt.xlabel("Activity")
plt.ylabel("Number of samples")

plt.show()
```
